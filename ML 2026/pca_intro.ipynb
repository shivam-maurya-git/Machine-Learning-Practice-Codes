```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
sns.pairplot(data = df)
plt.show()
```

```python
df.corr()
```

```python
model = PCA()
```

```python
pca_features = model.fit_transform(df)
```

```python
pca_features
```

```python
len(pca_features)
```

```python
df.shape
```

```python
df_pca = pd.DataFrame(pca_features)
```

```python
df_pca
```

```python
corr = df_pca.corr()
```

```python
corr
```

```python
sns.pairplot(data = df_pca)
plt.show()
```

```python
model.explained_variance_ratio_ #    Percentage of variance explained by each of the selected components.
```

```python
pca_var = model.explained_variance_
```

```python
model.n_components_
```

```python
plt.bar(x = [1,2,3,4,5,6,7,8,9],height = pca_var)
plt.summer()
plt.show() # amount of var explained by each feature
```

```python
pca_mean = model.mean_ #coordinates of the mean of the features
```
