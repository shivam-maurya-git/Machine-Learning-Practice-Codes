```python
import pandas as pd
from sklearn.decomposition import NMF
import numpy as np
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
model = NMF(n_components=3)
```

```python
df.shape
```

```python
transformed = model.fit_transform(df) # W _mtarix
```

```python
transformed
```

```python
transformed.shape # W matrix
```

```python
matrix_H = model.components_
```

```python
matrix_H
```

```python
matrix_H.shape
```

```python
df_comp = pd.DataFrame(matrix_H,columns = df.columns)
```

```python
df_comp
```

```python
for i in range(0,3):
    print(df_comp.loc[i].nlargest(3))
    print("\n")
    
```
