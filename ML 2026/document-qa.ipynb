```python
from pypdf import PdfReader
```

```python
reader = PdfReader("challenegs_in_teaching_laws_of_exponents.pdf")
```

```python
reader
```

```python
text = ""
for page in reader.pages:
    text +=page.extract_text()
    
```

```python
text
```

```python
from transformers import pipeline
```

```python
pipe = pipeline(task = "question-answering",model= "distilbert-base-cased-distilled-squad")
```

```python
question = "What is Misconception of generalization?"
```

```python
pipe(question =question, context=text )
```
